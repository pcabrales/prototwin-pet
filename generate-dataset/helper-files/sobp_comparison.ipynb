```python
import numpy as np
import matplotlib.pyplot as plt

activation = True  # CHANGE TO COMPARE ACTIVATION OR DOSE

if activation:
    img_type = "activation"
else:
    img_type = "dose"
    
img0 = np.load(f"/home/pablo/prototwin/deep-learning-dose-activity-dictionary/data/head-sobp-dataset0/{img_type}_uncropped/sobp0.npy")
img1 = np.load(f"/home/pablo/prototwin/deep-learning-dose-activity-dictionary/data/head-sobp-dataset2/{img_type}_uncropped/sobp0.npy")

l2_loss = np.sum((img0 - img1) ** 2)
print(l2_loss)

img0_profile = img0[:, img0.shape[1]//2 + 10, img0.shape[2]//2 + 10]
img1_profile = img1[:, img1.shape[1]//2 + 10, img1.shape[2]//2 + 10] 
# img0_profile = np.sum(img0, axis=(1, 2))
# img1_profile = np.sum(img1, axis=(1, 2))
distance = np.flip(np.arange(len(img0_profile)))

fig_name = f"./images/{img_type}_1e4_vs_1e5_head.png"

plt.plot(distance, img0_profile, label='1e5')
plt.plot(distance, img1_profile, label='1e4')
plt.xlabel('Depth (mm)')
plt.ylabel(f"{img_type} (a.u.)")
plt.legend()
plt.savefig(fig_name, dpi=500)
plt.show()
```
